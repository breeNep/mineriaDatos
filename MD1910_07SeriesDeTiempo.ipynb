```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
url = 'https://raw.githubusercontent.com/omarmendoza564/datos/main/datos/data_nybirths.dat'
df_births = pd.read_csv(url, header=None, names=['Births'])
df_births
```

```python
df_births.index = pd.date_range(start='1946-01', periods=len(df_births), freq='M')
df_births
```

```python
df_births.plot()
plt.title('New York Monthly Live Birth 1946-1959')
plt.xlabel('Year')
plt.ylabel('Number of birth')
plt.show()
```

```python
#Sacamos mes
df_births['Month'] = df_births.index.month
df_births
```

```python
plt.figure(figsize=(12,8))
df_births.boxplot(column='Births', by='Month')
plt.ylabel('Number of births')
#Nacimientoss registrados independientemente del mes
```

```python
#Tendencia(estacionalidad)no tenga un patron estable
plt.figure(figsize=(12,8))
decomposition = seasonal_decompose(df_births['Births'], model='additive', period=12)
decomposition.plot()
plt.show()
```

```python
#Cada elemento separado
plt.figure(figsize=(12,8))
plt.subplot(211)
plt.plot(decomposition.trend)
```
